```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
 genero_df = pd.read_csv("Generos_na_escola.csv")
```

```python
genero_df
```

```python
# Selecionando as colunas relevantes para o scatter plot
df_plot = genero_df[['1970', '1980', '1990', '2000']]

# Criando o scatter plot com 4 dimensões
plt.figure(figsize=(10, 6))
plt.scatter(
    df_plot['1970'],  # Eixo x
    df_plot['1980'],  # Eixo y
    s=df_plot['1990'] * 100,  # Tamanho das bolinhas
    c=df_plot['2000'],  # Cor das bolinhas
    cmap='viridis'  # Mapa de cores
)

# Adicionando rótulos aos eixos
plt.xlabel('Proporção de Gêneros na Escola em 1970')
plt.ylabel('Proporção de Gêneros na Escola em 1980')

# Adicionando um título ao gráfico
plt.title('Comparação da Proporção de Gêneros na Escola em 4 Décadas')

# Adicionando barra de cores para a dimensão "cor"
plt.colorbar(label='Proporção de Gêneros na Escola em 2000')

# Mostrando o gráfico
plt.show()
```
